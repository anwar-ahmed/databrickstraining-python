### Machine Learning Driving Business Value

Machine learning discovers patterns within data without being explicitly programmed.  As businesses adopt and utilize data in increasingly robust ways, machine learning offers the next step beyond just using data analytics to inform decision-making.  Use cases for machine learning include:<br><br>

* **Fraud Detection:** classifying real users from bad actors on a website
* **A/B Testing:** offering users different versions of a website indicates what the most successful version will be
* **Image Recognition:** medical image processing, self-driving cars, and facial recognition all use image recognition techniques
* **Natural Language Processing:** analyzing natural language using statistical methods allow for the algorithmic understanding of language, empowering sentiment analysis, chatbots, and many other applications
* **Financial Forecasting:** future company financials can be predicted using past data
* **Churn Analysis:** determine whether a customer might leave a business determines the best way to retain customers

Apache Spark allows us to perform machine learning tasks when data is too large to sit on any one machine.  In brief, Spark allows for scale beyond common approaches such as SciKit-Learn and R.  In addition to solving this technically challenging use case of distributed learning, Spark does this using an API that integrates with the rest of a production workflow.  This enables production-ready, machine learning jobs at scale in any Spark-driven pipeline.

## Summary
* Click the down arrow on a folder and select the **Create Notebook** option to create notebooks.
* Click the down arrow on a folder and select the **Import** option to import notebooks.
* Select the **Attached/Detached** option directly below the notebook title to attach to a spark cluster 
* Create clusters using the Clusters button on the left sidebar.

## Review

**Question:** How do you create a Notebook?  
**Answer:** Sign into Databricks, select the **Home** icon from the sidebar, right-click your home-folder, select **Create**, and then **Notebook**. In the **Create Notebook** dialog, specify the name of your notebook and the default programming language.

**Question:** How do you create a cluster?  
**Answer:** Select the **Clusters** icon on the sidebar, click the **Create Cluster** button, specify the specific settings for your cluster and then click **Create Cluster**.

**Question:** How do you attach a notebook to a cluster?  
**Answer:** If you run a command while detached, you may be prompted to connect to a cluster. To connect to a specific cluster, open the cluster menu by clicking the **Attached/Detached** menu item and then selecting your desired cluster.

**Question:** What does it mean to "attach" to a cluster?  
**Answer:** Attaching a notebook to a cluster means that the  current notebook will execute its code on that cluster. You could also say that when attached, the notebook is using the resources of that cluster.
